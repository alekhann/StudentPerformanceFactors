# Task
Build a linear regression model to predict student scores based on study hours using the provided DataFrame. Split the data into training and testing sets, train the model, and visualize the results.

## Prepare data

### Subtask:
Select the features and target variable for the linear regression model.


**Reasoning**:
Select the feature and target variables and reshape the feature variable for the linear regression model.



```python
X = df['Hours_Studied'].values.reshape(-1, 1)
y = df['Exam_Score']
```

## Split data

### Subtask:
Split the data into training and testing sets.


**Reasoning**:
Split the feature and target variables into training and testing sets using train_test_split.



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Build and train model

### Subtask:
Create and train a linear regression model using the training data.


**Reasoning**:
Import the LinearRegression model and train it using the training data.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

## Evaluate model

### Subtask:
Evaluate the performance of the model on the testing data.


**Reasoning**:
Import necessary metrics, make predictions, and calculate and print the evaluation scores.



```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2) Score: {r2}")
```

## Visualize results

### Subtask:
Create visualizations to show the relationship between study hours and exam scores, and the model's predictions.


**Reasoning**:
Generate a scatter plot of the test data and the linear regression line to visualize the model's performance.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression: Exam Score vs. Hours Studied')
plt.legend()
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The Mean Squared Error (MSE) of the model on the test data is approximately 12.18.
*   The R-squared (R2) score of the model on the test data is approximately 0.20.
*   A plot was generated showing the actual exam scores versus hours studied, along with the linear regression line predicting exam scores based on study hours.

### Insights or Next Steps

*   The low R-squared value (0.20) suggests that study hours alone do not explain a large proportion of the variance in exam scores. Other factors likely influence exam performance.
*   Further analysis could involve exploring additional features (e.g., attendance, previous grades, study environment) to build a more predictive model.
